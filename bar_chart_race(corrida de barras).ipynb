#Mortes por estados

```
cases_state = 'https://raw.githubusercontent.com/wcota/covid19br/master/cases-brazil-states.csv'
```

```
estados = pd.read_csv(cases_state, 
                      sep=',', 
                      encoding='UTF-8', 
                      index_col='date', 
                      parse_dates=['date'])
```

```
estados.head()
```

```
estados.loc[estados['state']=='PI', :]
```

```
estados.groupby(['state', 'date']).sum()['deaths']
```

```
pd.DataFrame(estados.groupby(['state', 'date']).sum()['deaths']).unstack().T
```

```
mortesEstados = pd.DataFrame(estados.groupby(['state', 'date']).sum()['deaths']).unstack().T.droplevel(level=0)
```

```
mortesEstados.tail(2)
```

```
bcr.bar_chart_race(mortesEstados, 
                   period_fmt='%-d/%m/%Y', 
                   steps_per_period=5, 
                   period_length=500, 
                   title='Mortes por COVID-19 (estados)',
                   figsize=(6,3))
```

```
# Gravando em disco
bcr.bar_chart_race(mortesEstados, 
                    'drive/My Drive/python/bcrCovid19.mp4', 
                    period_fmt='%-d %b %Y', 
                    steps_per_period=5, 
                    period_length=500,
                    title='Mortes por COVID-19 (estados) ', 
                    figsize=(5, 3))
```

```
casosEstados = pd.DataFrame(estados.groupby(['state', 'date']).sum()['totalCases']).unstack().T.droplevel(level=0)
```

```
casosEstados.head()
```

```
bcr.bar_chart_race(casosEstados, 
                   period_fmt='%d %b %Y', 
                   steps_per_period=5, 
                   period_length=500, 
                   title='Casos confirmados de COVID-19 (estados)')
```

```
# Gravando em disco
bcr.bar_chart_race(casosEstados, 
                    'drive/My Drive/python/casosEstados.mp4', 
                    period_fmt='%-d %b %Y', 
                    steps_per_period=5, 
                    period_length=500,
                    title='Mortes por COVID-19 por estados ', 
                    figsize=(5, 3))
```

#Casos por cidades piauienses

```
file_cid = 'https://raw.githubusercontent.com/wcota/covid19br/master/cases-brazil-cities-time.csv'
```

```
cidades = pd.read_csv(file_cid, 
                      sep=',', 
                      encoding='UTF-8', 
                      parse_dates=['date'])
```

```
cidades.tail()
```

```
# Selecionando as cidades do PI
cidades_PI = cidades.loc[cidades['state']=='PI', :] 
```

```
cidades_PI.head()
```

```
# Seleciona as linhas que esto sem dados de localizao
df_remv = cidades_PI.loc[cidades_PI['city']=='CASO SEM LOCALIZAO DEFINIDA/PI', : ]
```

```
df_remv.head()
```

```
# Remove as linhas selecinadas anteriormente
cidades_PI = cidades_PI.drop(df_remv.index)
```

```
# Agrupa o dataframe por cidade e data e faz uma somatria dos casos totais
cidades_PI.groupby(['city', 'date']).sum()['totalCases'].unstack().T
```

```
# Cria um novo df agrupado por cidade e data e com a somatria dos casos totais 
casos_cidades = pd.DataFrame(cidades_PI.groupby(['city', 'date']).sum()['totalCases']).unstack().T.droplevel(level=0)
```

```
casos_cidades.head()
```

```
bcr.bar_chart_race(casos_cidades, 
                   period_fmt='%-d %m %Y', 
                   title='Casos de COVID-19 por cidades piauienses (Top 10)', 
                   steps_per_period=5, 
                   period_length=500,
                   n_bars=10)
```

```
bcr.bar_chart_race(casos_cidades, 
                   'drive/My Drive/python/casosCid.mp4', 
                   period_fmt='%-d %b %Y', 
                   title='Casos de COVID-19 por cidades piauienses (Top 10)', 
                   steps_per_period=5, 
                   period_length=500,
                   n_bars=10,
                   figsize=(5,3))
```
